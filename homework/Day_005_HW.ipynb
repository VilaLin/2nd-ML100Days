```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
app_train.head()
```

```python
print(app_train.head()['AMT_INCOME_TOTAL'])
print(app_train.head()['AMT_INCOME_TOTAL'].max())
print(app_train.head()['AMT_INCOME_TOTAL'].std())
print(app_train.head()['AMT_INCOME_TOTAL'].mean())
print(app_train.head()['AMT_INCOME_TOTAL'].sum())
print(app_train.head()['AMT_INCOME_TOTAL'].describe())
```

```python
plt.hist(x=app_train.head(10)['AMT_INCOME_TOTAL'])
plt.title('AMT_INCOME_TOTAL bar chart')
plt.xlabel('Value')
plt.ylabel('No.')
plt.show()
```
